```python
import json
import csv
```

```python
# Ruta completa al archivo JSON
json_file = "dir of json file"

# Ruta completa donde quieres guardar el CSV
csv_file = "dir for csv file "
```

```python
# Carga los datos del JSON
with open(json_file, 'r') as data:
    json_data = json.load(data)

# Extrae los encabezados de la primera fila
headers = json_data[0].keys()

# Escribe los datos en el CSV
with open(csv_file, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers)
    writer.writeheader()
    for row in json_data:
        writer.writerow(row)

print("CSV succesfully created in:", csv_file)
```
